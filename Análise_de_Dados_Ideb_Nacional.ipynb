## Obtendo tabelas separadas de acordo com a rede (pública ou privada), ensino e ano escolar

```python
#Tabela com os índices das escolas públicas dos anos iniciais do ensino fundamental durante todo o período de 2005 à 2021
df_pub_fund_inic = df_brasil[((df_brasil["rede"] == "publica") | (df_brasil["rede"] == "pública")) & (df_brasil["ensino"] == "fundamental") & (df_brasil["anos_escolares"] == "iniciais (1-5)")]
```

```python
#Tabela com os índices das escolas públicas dos anos finais do ensino fundamental durante todo o período de 2005 à 2021
df_pub_fund_fin = df_brasil[((df_brasil["rede"] == "publica") | (df_brasil["rede"] == "pública")) & (df_brasil["ensino"] == "fundamental") & (df_brasil["anos_escolares"] == "finais (6-9)")]
```

```python
#Tabela com os índices das escolas públicas de todos os anos do ensino médio durante todo o período de 2005 à 2021
df_pub_medio_todos = df_brasil[((df_brasil["rede"] == "publica") | (df_brasil["rede"] == "pública")) & (df_brasil["ensino"] == "medio")]
```

```python
#Tabela com os índices das escolas privadas dos anos iniciais do ensino fundamental durante todo o período de 2005 à 2021
df_priv_fund_inic = df_brasil[(df_brasil["rede"] == "privada") & (df_brasil["ensino"] == "fundamental") & (df_brasil["anos_escolares"] == "iniciais (1-5)")]
```

```python
#Tabela com os índices das escolas privadas dos anos finais do ensino fundamental durante todo o período de 2005 à 2021
df_priv_fund_fin = df_brasil[(df_brasil["rede"] == "privada") & (df_brasil["ensino"] == "fundamental") & (df_brasil["anos_escolares"] == "finais (6-9)")]
```

```python
#Tabela com os índices das escolas privadas de todos os anos do ensino médio durante todo o período de 2005 à 2021
df_priv_medio_todos = df_brasil[(df_brasil["rede"] == "privada") & (df_brasil["ensino"] == "medio") & (df_brasil["anos_escolares"] == "todos (1-4)")]
```

## Plotando gráficos de cada um dos dados selecionados

```python
def plotar1(data, x, y, titulo, nome):
    fig, ax = plt.subplots(figsize = (16,8))
    ax.plot(data[x], data[y])
    plt.title(titulo, fontsize=18)
    plt.xticks(range(2005, 2022, 2))
    ax.set_xlabel("Ano", size = 16)
    ax.set_ylabel("Ideb", size = 16)
    plt.savefig("C:\\Users\\igord\\OneDrive\\Área de Trabalho\\PS Analytica\\plots\\" + nome, dpi=300)
    plt.show()
```

```python
plotar1(df_pub_fund_inic, "ano", "ideb", "Evolução do Ideb nas escolas públicas nos anos iniciais do ensino fundamental", "pub_fund_inic")
```

```python
plotar1(df_pub_fund_fin, "ano", "ideb", "Evolução do Ideb nas escolas públicas nos anos finais do ensino fundamental","pub_fund_fin")
```

```python
plotar1(df_pub_medio_todos, "ano", "ideb", "Evolução do Ideb nas escolas públicas no ensino médio","pub_medio_todos")
```

```python
plotar1(df_priv_fun_inic,"ano", "ideb","Evolução do Ideb nas escolas privadas nos anos iniciais do ensino fundamental", "priv_fund_inic")
```

```python
plotar1(df_priv_fun_fin,"ano", "ideb","Evolução do Ideb nas escolas privadas nos anos finais do ensino fundamental", "priv_fund_fin")
```

```python
plotar1(df_priv_medio_todos,"ano", "ideb","Evolução do Ideb nas escolas privadas no ensino médio", "priv_medio_todos")
```

```python
def plotar2(lista, x, y, titulo, nome):
    fig, ax = plt.subplots(figsize = (16,8))
    for data in lista:
        ax.plot(data[x], data[y])
    plt.title(titulo, fontsize=18)
    plt.xticks(range(2005, 2022, 2))
    ax.legend(["Públicas", "Privadas"], loc = "lower right", fontsize = 15)
    ax.set_xlabel("Ano", size = 16)
    ax.set_ylabel("Ideb", size = 16)
    plt.savefig("C:\\Users\\igord\\OneDrive\\Área de Trabalho\\PS Analytica\\plots\\" + nome, dpi=300)
    plt.show()
```

```python
plotar2([df_pub_fund_inic, df_priv_fun_inic], "ano", "ideb", "Evolução do Ideb nas escola públicas e privadas nos anos iniciais do ensino fundamental", "df_pub_priv_fund_inic")
```

```python
plotar2([df_pub_fund_fin, df_priv_fun_fin], "ano", "ideb", "Evolução do Ideb nas escola públicas e privadas nos anos finais do ensino fundamental", "df_pub_priv_fund_fin")
```

```python
plotar2([df_pub_medio_todos, df_priv_medio_todos], "ano", "ideb", "Evolução do Ideb nas escola públicas e privadas durante o ensino médio", "df_pub_priv_medio_todos")
```

```python
#Lista com data frames para facilitar a iteração
lista_tabelas = [df_pub_fund_inic, df_pub_fund_fin, df_pub_medio_todos, df_priv_fund_inic, df_priv_fund_fin, df_priv_medio_todos]
```

## Dá para notar pelo crescimento relativo que as escolas públicas obtiveram uma melhoria considerável no índice, mas ainda há uma diferença susbtâncial entre as escola públicas e privadas

```python
for tab in lista_tabelas:
    print((tab["ideb"].iloc[-1] - tab["ideb"].iloc[0]) / tab["ideb"].iloc[0])
```

```python
lista_dif = [[],[],[]]
```

```python
# Obtendo as diferenças em cada etapa da educação, subtraindo o ideb da rede pública do ideb da rede privada

for n,lista in enumerate(lista_dif):
    for i in range(9):
        lista.append(round(lista_tabelas[n+3]["ideb"].iloc[i] - lista_tabelas[n]["ideb"].iloc[i], 1))
```

```python
lista_dif
```

```python
fig, ax = plt.subplots(figsize = (16,8))
for l in lista_dif:
    ax.plot(range(2005, 2022, 2),l)
plt.title("Diferença entre os índices das escolas privadas e públicas", fontsize=18)
plt.xticks(range(2005, 2022, 2))
ax.legend(["Anos iniciais fundamental", "Anos finais fundamental", "Todos anos médio"], loc = "upper right", fontsize = 15)
ax.set_xlabel("Ano", size = 16)
ax.set_ylabel("Diferença Ideb", size = 16)
ax.grid(True)
plt.savefig("C:\\Users\\igord\\OneDrive\\Área de Trabalho\\PS Analytica\\plots\\dif_ideb", dpi=300)
plt.show()
```
